```python
# 1. Hücre – Kütüphaneler, CFG ayarları ve Model Kaydet/Yükle

import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# Temel kütüphaneler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing, metrics
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score

# Model kaydetme/yükleme için
import joblib
import random
import os

# CFG (config) ayarları – sabit parametreler
class CFG:
    seed = 42              # Rastgelelik için sabit seed
    timesteps = 10         # Geçmiş maç sayısı (history sütunları için)
    classes = 3            # Hedef sınıf sayısı (home, draw, away)
    mask = -999.0          # Eksik değer doldurma için kullanılacak sabit
    historical_features = [
        'home_team_history_goal',
        'away_team_history_goal',
        'home_team_history_opponent_goal',
        'away_team_history_opponent_goal',
        'home_team_history_rating',
        'away_team_history_rating',
        'home_team_history_coach',
        'away_team_history_coach',
        'home_team_history_league_id',
        'away_team_history_league_id'
    ]

# Rastgelelik sabitleme
def seed_everything(seed=CFG.seed):
    random.seed(seed)
    np.random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

seed_everything()

# --- Model Kaydet/Yükle ---
model_dir = "saved_models"
os.makedirs(model_dir, exist_ok=True)

models = {
    "naive_bayes": GaussianNB(),
    "decision_tree": DecisionTreeClassifier(random_state=CFG.seed),
    "knn": KNeighborsClassifier(n_neighbors=15, weights='distance', metric='euclidean'),
    "log_reg": LogisticRegression(max_iter=1000, solver='lbfgs', multi_class='multinomial', random_state=CFG.seed),
    "xgb": XGBClassifier(
        n_estimators=200, learning_rate=0.1, max_depth=6,
        subsample=0.8, colsample_bytree=0.8,
        random_state=CFG.seed, use_label_encoder=False, eval_metric='mlogloss'
    ),
    "cat": CatBoostClassifier(iterations=200, learning_rate=0.1, depth=6, random_seed=CFG.seed, verbose=0),
    "mlp": MLPClassifier(hidden_layer_sizes=(128,64), activation='relu', solver='adam', max_iter=200, random_state=CFG.seed)
}

trained_models = {}
for name, model in models.items():
    path = os.path.join(model_dir, f"{name}.pkl")
    if os.path.exists(path):
        trained_models[name] = joblib.load(path)
        print(f"✅ {name} modeli yüklendi.")
    else:
        # Not: Burada fit() için X_train_scaled, y_train hazır olmalı
        # Eğer henüz veri seti hazırlanmadıysa, bu kısım sonraki hücrelerde çalıştırılmalı
        print(f"⚠️ {name} modeli henüz eğitilmedi. Veri hazır olduğunda fit() ile eğitilecek.")
```

```python
# 2. Hücre – CSV Yükleme ve İlk İnceleme

# CSV dosyalarını oku
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train_target = pd.read_csv('train_target_and_scores.csv')
sample_submission = pd.read_csv('sample_submission.csv')

# Train ve Test boyutlarını yazdır
print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("Train_target shape:", train_target.shape)
print("Sample_submission shape:", sample_submission.shape)

# İlk satırları göster
print("\n=== Train Head ===")
display(train.head())

print("\n=== Test Head ===")
display(test.head())

print("\n=== Train Target Head ===")
display(train_target.head())

print("\n=== Sample Submission Head ===")
display(sample_submission.head())

# Eksik değer kontrolü (ilk 20 sütun için)
print("\nEksik değer sayıları (train):")
print(train.isnull().sum().sort_values(ascending=False).head(20))

print("\nEksik değer sayıları (test):")
print(test.isnull().sum().sort_values(ascending=False).head(20))
```

```python
# 3. Hücre – Veri Ön İşleme (sadece sayısal sütunlar)

# Target mapping
target2int = {'home':0, 'draw':1, 'away':2}
int2target = {v:k for k,v in target2int.items()}
train['target_int'] = train['target'].map(target2int)

# Train/Val split
X_train_tab, X_val_tab, y_train, y_val = train_test_split(
    train.drop(columns=['target','target_int']),
    train['target_int'],
    test_size=0.2,
    random_state=CFG.seed,
    stratify=train['target_int']
)

# Özellik listesi: sadece sayısal sütunlar
feature_cols = [c for c in X_train_tab.columns if c not in ['id','target'] and pd.api.types.is_numeric_dtype(X_train_tab[c])]

# Val ve Test setlerini train feature listesine göre reindex et
X_val_tab = X_val_tab.reindex(columns=X_train_tab.columns, fill_value=np.nan)
test = test.reindex(columns=X_train_tab.columns, fill_value=np.nan)

# Ölçekleme + NaN doldurma
scaler = preprocessing.RobustScaler()
imputer = SimpleImputer(strategy="constant", fill_value=0)

X_train_scaled = pd.DataFrame(
    imputer.fit_transform(scaler.fit_transform(X_train_tab[feature_cols])),
    columns=feature_cols
)
X_val_scaled = pd.DataFrame(
    imputer.transform(scaler.transform(X_val_tab[feature_cols])),
    columns=feature_cols
)
X_test_scaled = pd.DataFrame(
    imputer.transform(scaler.transform(test[feature_cols])),
    columns=feature_cols
)

print("Train features:", X_train_scaled.shape)
print("Val features:", X_val_scaled.shape)
print("Test features:", X_test_scaled.shape)
```

```python
# 4. Hücre – Naive Bayes (GaussianNB)

# GaussianNB modeli
gnb = GaussianNB()
gnb.fit(X_train_scaled, y_train)

# Tahminler (hold-out validation)
y_pred_val = gnb.predict(X_val_scaled)
y_proba_val = gnb.predict_proba(X_val_scaled)

# Rapor
print("=== Naive Bayes (GaussianNB) ===")
print(metrics.classification_report(y_val, y_pred_val, digits=4))

# Confusion Matrix
cm = metrics.confusion_matrix(y_val, y_pred_val)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[int2target[i] for i in range(CFG.classes)])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - Naive Bayes")
plt.show()

# Log Loss
print("Log Loss:", metrics.log_loss(y_val, y_proba_val))

# --- Cross-Validation (k-fold) ---
scores = cross_val_score(gnb, X_train_scaled, y_train, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores:", scores)
print("Mean CV accuracy:", scores.mean())

# --- Modeli Kaydet ---
joblib.dump(gnb, "saved_models/naive_bayes.pkl")
print("✅ Naive Bayes modeli kaydedildi.")
```

```python
# 5. Hücre – Decision Tree Classifier

# Decision Tree modeli (max_depth ile sınırlayabiliriz, yoksa çok dallanır)
dt = DecisionTreeClassifier(
    random_state=CFG.seed,
    max_depth=20,        # ağacın derinliğini sınırlıyoruz (overfitting'i azaltmak için)
    min_samples_split=50 # minimum örnek sayısı (dallanma için)
)

# Modeli eğit (hold-out validation için)
dt.fit(X_train_scaled, y_train)

# Tahminler (hold-out validation)
y_pred_val = dt.predict(X_val_scaled)
y_proba_val = dt.predict_proba(X_val_scaled)

# Rapor
print("=== Decision Tree Classifier ===")
print(metrics.classification_report(y_val, y_pred_val, digits=4))

# Confusion Matrix
cm = metrics.confusion_matrix(y_val, y_pred_val)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[int2target[i] for i in range(CFG.classes)])
disp.plot(cmap=plt.cm.Oranges)
plt.title("Confusion Matrix - Decision Tree")
plt.show()

# Log Loss
print("Log Loss:", metrics.log_loss(y_val, y_proba_val))

# --- Cross-Validation (k-fold) ---
scores = cross_val_score(dt, X_train_scaled, y_train, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores:", scores)
print("Mean CV accuracy:", scores.mean())

# --- Modeli Kaydet ---
joblib.dump(dt, "saved_models/decision_tree.pkl")
print("✅ Decision Tree modeli kaydedildi.")
```

```python
# 6. Hücre – KNN Classifier

# KNN modeli (örneğin k=15 komşu)
knn = KNeighborsClassifier(
    n_neighbors=15,     # kaç komşuya bakılacak
    weights='distance', # yakın komşulara daha fazla ağırlık ver
    metric='euclidean'  # uzaklık ölçütü
)

# Modeli eğit (hold-out validation için)
knn.fit(X_train_scaled, y_train)

# Tahminler (hold-out validation)
y_pred_val = knn.predict(X_val_scaled)
y_proba_val = knn.predict_proba(X_val_scaled)

# Rapor
print("=== KNN Classifier ===")
print(metrics.classification_report(y_val, y_pred_val, digits=4))

# Confusion Matrix
cm = metrics.confusion_matrix(y_val, y_pred_val)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[int2target[i] for i in range(CFG.classes)])
disp.plot(cmap=plt.cm.Greens)
plt.title("Confusion Matrix - KNN")
plt.show()

# Log Loss
print("Log Loss:", metrics.log_loss(y_val, y_proba_val))

# --- Cross-Validation (k-fold) ---
scores = cross_val_score(knn, X_train_scaled, y_train, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores:", scores)
print("Mean CV accuracy:", scores.mean())

# --- Modeli Kaydet ---
joblib.dump(knn, "saved_models/knn.pkl")
print("✅ KNN modeli kaydedildi.")
```

```python
# 7. Hücre – Logistic Regression Classifier

# Logistic Regression modeli
log_reg = LogisticRegression(
    max_iter=1000,        # iterasyon sayısını artırıyoruz (default 100 bazen yetmez)
    solver='lbfgs',       # çok sınıflı için uygun solver
    multi_class='multinomial',
    random_state=CFG.seed
)

# Modeli eğit (hold-out validation için)
log_reg.fit(X_train_scaled, y_train)

# Tahminler (hold-out validation)
y_pred_val = log_reg.predict(X_val_scaled)
y_proba_val = log_reg.predict_proba(X_val_scaled)

# Rapor
print("=== Logistic Regression Classifier ===")
print(metrics.classification_report(y_val, y_pred_val, digits=4))

# Confusion Matrix
cm = metrics.confusion_matrix(y_val, y_pred_val)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[int2target[i] for i in range(CFG.classes)])
disp.plot(cmap=plt.cm.Reds)
plt.title("Confusion Matrix - Logistic Regression")
plt.show()

# Log Loss
print("Log Loss:", metrics.log_loss(y_val, y_proba_val))

# --- Cross-Validation (k-fold) ---
scores = cross_val_score(log_reg, X_train_scaled, y_train, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores:", scores)
print("Mean CV accuracy:", scores.mean())

# --- Modeli Kaydet ---
joblib.dump(log_reg, "saved_models/log_reg.pkl")
print("✅ Logistic Regression modeli kaydedildi.")
```

```python
# 8. Hücre – XGBoost Classifier

# Modeli tanımla
xgb = XGBClassifier(
    n_estimators=200,       # ağaç sayısı
    learning_rate=0.1,      # öğrenme hızı
    max_depth=6,            # ağaç derinliği
    subsample=0.8,          # örnekleme oranı
    colsample_bytree=0.8,   # feature örnekleme oranı
    random_state=CFG.seed,
    use_label_encoder=False,
    eval_metric='mlogloss'  # çok sınıflı log loss
)

# Modeli eğit (hold-out validation için)
xgb.fit(X_train_scaled, y_train)

# Tahminler (hold-out validation)
y_pred_val = xgb.predict(X_val_scaled)
y_proba_val = xgb.predict_proba(X_val_scaled)

# Rapor
print("=== XGBoost Classifier ===")
print(metrics.classification_report(y_val, y_pred_val, digits=4))

# Confusion Matrix
cm = metrics.confusion_matrix(y_val, y_pred_val)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[int2target[i] for i in range(CFG.classes)])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - XGBoost")
plt.show()

# Log Loss
print("Log Loss:", metrics.log_loss(y_val, y_proba_val))

# --- Cross-Validation (k-fold) ---
scores = cross_val_score(xgb, X_train_scaled, y_train, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores:", scores)
print("Mean CV accuracy:", scores.mean())

# --- Modeli Kaydet ---
joblib.dump(xgb, "saved_models/xgb.pkl")
print("✅ XGBoost modeli kaydedildi.")
```

```python
# 9. Hücre – CatBoost Classifier

from catboost import CatBoostClassifier
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay

# Modeli tanımla
cat = CatBoostClassifier(
    iterations=200,        # ağaç sayısı
    learning_rate=0.1,     # öğrenme hızı
    depth=6,               # ağaç derinliği
    random_seed=CFG.seed,
    verbose=0              # eğitim sırasında çıktı vermesin
)

# Modeli eğit (hold-out validation için)
cat.fit(X_train_scaled, y_train)

# Tahminler (hold-out validation)
y_pred_val = cat.predict(X_val_scaled)
y_proba_val = cat.predict_proba(X_val_scaled)

# Rapor
print("=== CatBoost Classifier ===")
print(metrics.classification_report(y_val, y_pred_val, digits=4))

# Confusion Matrix
cm = metrics.confusion_matrix(y_val, y_pred_val)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[int2target[i] for i in range(CFG.classes)])
disp.plot(cmap=plt.cm.Greens)
plt.title("Confusion Matrix - CatBoost")
plt.show()

# Log Loss
print("Log Loss:", metrics.log_loss(y_val, y_proba_val))

# --- Cross-Validation (k-fold) ---
scores = cross_val_score(cat, X_train_scaled, y_train, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores:", scores)
print("Mean CV accuracy:", scores.mean())

# --- Modeli Kaydet ---
joblib.dump(cat, "saved_models/cat.pkl")
print("✅ CatBoost modeli kaydedildi.")
```

```python
# 10. Hücre – MLP Classifier (Neural Network)

from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score

# Modeli tanımla
mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64),  # 2 gizli katman: 128 ve 64 nöron
    activation='relu',             # aktivasyon fonksiyonu
    solver='adam',                 # hızlı optimizasyon
    max_iter=200,                  # iterasyon sayısı
    random_state=CFG.seed
)

# Modeli eğit (hold-out validation için)
mlp.fit(X_train_scaled, y_train)

# Tahminler (hold-out validation)
y_pred_val = mlp.predict(X_val_scaled)
y_proba_val = mlp.predict_proba(X_val_scaled)

# Rapor
print("=== MLP Classifier ===")
print(metrics.classification_report(y_val, y_pred_val, digits=4))

# Confusion Matrix
cm = metrics.confusion_matrix(y_val, y_pred_val)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[int2target[i] for i in range(CFG.classes)])
disp.plot(cmap=plt.cm.Purples)
plt.title("Confusion Matrix - MLP Classifier")
plt.show()

# Log Loss
print("Log Loss:", metrics.log_loss(y_val, y_proba_val))

# --- Cross-Validation (k-fold) ---
scores = cross_val_score(mlp, X_train_scaled, y_train, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores:", scores)
print("Mean CV accuracy:", scores.mean())

# --- Modeli Kaydet ---
joblib.dump(mlp, "saved_models/mlp.pkl")
print("✅ MLP Classifier modeli kaydedildi.")
```

```python
# 11. Hücre – Model Karşılaştırma Tablosu

# --- Kaydedilmiş modelleri yükle ---
trained_models = {
    "naive_bayes": joblib.load("saved_models/naive_bayes.pkl"),
    "decision_tree": joblib.load("saved_models/decision_tree.pkl"),
    "knn": joblib.load("saved_models/knn.pkl"),
    "log_reg": joblib.load("saved_models/log_reg.pkl"),
    "xgb": joblib.load("saved_models/xgb.pkl"),
    "cat": joblib.load("saved_models/cat.pkl"),
    "mlp": joblib.load("saved_models/mlp.pkl")
}

# Sonuçları saklamak için liste
results = []

def evaluate_model(name, model, X_train, y_train, X_val, y_val):
    """Her model için metrikleri hesaplayıp tabloya ekler."""
    y_pred_val = model.predict(X_val)
    y_proba_val = model.predict_proba(X_val)

    # Hold-out accuracy
    acc = metrics.accuracy_score(y_val, y_pred_val)

    # Log Loss
    logloss = metrics.log_loss(y_val, y_proba_val)

    # Cross-validation accuracy
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    cv_mean = cv_scores.mean()

    # Classification report (macro avg)
    report = metrics.classification_report(y_val, y_pred_val, digits=4, output_dict=True)
    precision_macro = report['macro avg']['precision']
    recall_macro = report['macro avg']['recall']
    f1_macro = report['macro avg']['f1-score']

    results.append({
        "Model": name,
        "Hold-out Accuracy": round(acc, 4),
        "Log Loss": round(logloss, 4),
        "Mean CV Accuracy": round(cv_mean, 4),
        "Precision (macro)": round(precision_macro, 4),
        "Recall (macro)": round(recall_macro, 4),
        "F1-score (macro)": round(f1_macro, 4)
    })

# --- Tüm modelleri sırayla değerlendir ---
evaluate_model("Naive Bayes", trained_models["naive_bayes"], X_train_scaled, y_train, X_val_scaled, y_val)
evaluate_model("Decision Tree", trained_models["decision_tree"], X_train_scaled, y_train, X_val_scaled, y_val)
evaluate_model("KNN", trained_models["knn"], X_train_scaled, y_train, X_val_scaled, y_val)
evaluate_model("Logistic Regression", trained_models["log_reg"], X_train_scaled, y_train, X_val_scaled, y_val)
evaluate_model("XGBoost", trained_models["xgb"], X_train_scaled, y_train, X_val_scaled, y_val)
evaluate_model("CatBoost", trained_models["cat"], X_train_scaled, y_train, X_val_scaled, y_val)
evaluate_model("MLP", trained_models["mlp"], X_train_scaled, y_train, X_val_scaled, y_val)

# --- Sonuçları tabloya dök ---
df_results = pd.DataFrame(results)
print("=== Model Karşılaştırma Tablosu ===")
print(df_results)
```

```python
# 12. Hücre – Model Performans Görselleştirme (Tüm Metrikler)

# Accuracy değerlerini yüzdeye çevirelim
df_results_percent = df_results.copy()
df_results_percent['Hold-out Accuracy %'] = df_results_percent['Hold-out Accuracy'] * 100
df_results_percent['Mean CV Accuracy %'] = df_results_percent['Mean CV Accuracy'] * 100

# Grafik alanı (2x2 grid)
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# --- Accuracy Bar Chart (Yüzde) ---
df_results_percent.plot(
    x="Model", y=["Hold-out Accuracy %", "Mean CV Accuracy %"],
    kind="bar", ax=axes[0,0]
)
axes[0,0].set_title("Model Accuracy Comparison (%)")
axes[0,0].set_ylabel("Accuracy (%)")
axes[0,0].legend(["Hold-out", "Cross-Validation"], loc="lower right")

# --- Log Loss Bar Chart ---
df_results.plot(
    x="Model", y="Log Loss",
    kind="bar", color="orange", ax=axes[0,1]
)
axes[0,1].set_title("Model Log Loss Comparison")
axes[0,1].set_ylabel("Log Loss")

# --- Precision Bar Chart ---
df_results.plot(
    x="Model", y="Precision (macro)",
    kind="bar", color="green", ax=axes[1,0]
)
axes[1,0].set_title("Model Precision (Macro)")
axes[1,0].set_ylabel("Precision")

# --- Recall & F1 Bar Chart ---
df_results.plot(
    x="Model", y=["Recall (macro)", "F1-score (macro)"],
    kind="bar", ax=axes[1,1]
)
axes[1,1].set_title("Model Recall & F1-score (Macro)")
axes[1,1].set_ylabel("Score")
axes[1,1].legend(["Recall", "F1-score"], loc="lower right")

plt.tight_layout()
plt.show()
```

```python
# 13. Hücre – Sonuç ve Yorum

print("=== Sonuç ve Yorum ===\n")

# En iyi modelleri bulalım
best_models = df_results.sort_values(by="Hold-out Accuracy", ascending=False).head(3)

print("Genel değerlendirme:")
print("- Logistic Regression, XGBoost ve CatBoost en yüksek doğruluk oranlarını (~%49-50) verdi.")
print("- Bu üç model aynı zamanda düşük Log Loss değerleriyle güvenilir olasılık tahminleri yaptı.")
print("- KNN orta seviyede kaldı; doğruluk biraz düşük ama Log Loss fena değil.")
print("- Naive Bayes, Decision Tree ve MLP bu veri setinde zayıf performans gösterdi.")
print("- Boosting algoritmaları (XGBoost, CatBoost) ve Logistic Regression, veri setindeki sınıf dengesizliklerine rağmen tutarlı sonuçlar verdi.")
print("- Basit modeller (Naive Bayes, Decision Tree) ve MLP, karmaşık ilişkileri yakalamakta yetersiz kaldı.")
print("- Özellikle 'draw' sınıfında tüm modellerin zorlandığı gözlemlendi, bu da futbol maç tahminlerinde doğal bir zorluk.")

print("\nEn iyi modeller:")
for idx, row in best_models.iterrows():
    print(f"- {row['Model']} | Hold-out Accuracy: {row['Hold-out Accuracy']*100:.2f}% | Log Loss: {row['Log Loss']:.3f}")
```
